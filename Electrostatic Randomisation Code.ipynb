```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

# Circle

```python
x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)

a, b = np.meshgrid(x,y)
circle = (a)**2+(b)**2-1

c = []
d = []
for i in range(20):
    r = np.sqrt(np.random.uniform(0,1))
    theta = np.random.uniform(0,2*np.pi)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    c.append(x)
    d.append(y)
    
fig, ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(6,6)
ax.contour(a,b,circle,[0])
ax.scatter(c,d,color='black')
ax.set_aspect(1)
plt.xlim(-1.1,1.1)
plt.ylim(-1.1,1.1)
plt.grid(linestyle='dotted')
```

```python
points = []
for i,j in zip(c,d):
    g = []
    g.append(i)
    g.append(j)
    points.append(g)
    list1 = []
for i in points:
    list2 = []
    g = [j for j in points if j !=i]
    for j in g:
        z = np.sqrt((j[0]-i[0])**2+(j[1]-i[1])**2)
        list2.append(np.log(z))
    mysum = -np.sum(list2)
    list1.append(mysum)
list1 = [i for i in list1 if i != 0]
for i,j in zip(points,list1):
    i.append(j)
```

```python
count = 0
condition = True
while condition == True:
    accepted=0
    count+=1
    f1 = []
    for i in points:
        f1.append(i[0])
        f1.append(i[1])
        f1.append(i[2])
    for time in range(1000):
        random.shuffle(points)
        if accepted > 50:
            dx = np.random.uniform(-(np.sqrt(2)/20),(np.sqrt(2)/20))
            dy = np.random.uniform(-(np.sqrt(2)/20),(np.sqrt(2)/20))
        else:
            dx = np.random.uniform(-(np.sqrt(2)/200),(np.sqrt(2)/200))
            dy = np.random.uniform(-(np.sqrt(2)/200),(np.sqrt(2)/200))
        i = points[0]
        ix = i[0]+dx
        iy = i[1]+dy
        g = [k for k in points if k!=i]
        list2 = []
        for k in g:
            z = np.sqrt((k[0]-ix)**2+(k[1]-iy)**2)
            list2.append(np.log(z))
        mysum = -np.sum(list2)
        if mysum<i[2] and -1<ix<1 and -1<iy<1 and ((ix)**2+(iy)**2)<1:
            i[0] = ix
            i[1] = iy
            i[2] = mysum
            accepted+=1
        else:
            i[0] = i[0]
            i[1] = i[1]
            i[2] = i[2]
        f2 = []
        for i in points:
            f2.append(i[0])
            f2.append(i[1])
            f2.append(i[2])
    print("On loop",count,"there were",accepted,"accepted moves.")
    if set(f1) == set(f2):
        condition = False
    else:
        condition = True
```

```python
g = []
h = []
for i in points:
    g.append(i[0])
    h.append(i[1])
    x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)

a, b = np.meshgrid(x,y)
circle = (a)**2+(b)**2-1

fig, ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(6,6)
ax.contour(a,b,circle,[0])
ax.scatter(g,h,color='black')
ax.set_aspect(1)
plt.xlim(-1.1,1.1)
plt.ylim(-1.1,1.1)
plt.grid(linestyle='dotted')
```

```python
gh = []
polarangles = []
for i,j in zip(g,h):
    k = [i,j]
    gh.append(k)
for i in gh:
    angle = np.arctan(i[1]/i[0])
    if i[0]<0 and i[1]>0:
        angle = angle+np.pi
    elif i[0]<0 and i[1]<0:
        angle = angle+np.pi
    elif i[0]>0 and i[1]<0:
        angle=angle+2*np.pi
    polarangles.append(angle)
print(np.sort(polarangles))
```

```python
s = np.sort(polarangles)
s1 = np.ndarray.tolist(s)
s1.append(s1[0]+2*np.pi)
s2 = np.diff(s1)
plt.hist(s2,bins='auto')
plt.grid(ls="--",axis="y")
plt.xlabel('Separation')
plt.ylabel('Frequency')
```
